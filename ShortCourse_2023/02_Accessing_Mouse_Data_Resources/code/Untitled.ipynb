```python
import requests #Required for interacting with MPD and GN
import json #Required for with JSON formatted data
from pandas.io.json import json_normalize #Required for indexing or calling a specific JSON variable

import io #This helps us in converting text based results from GN into a data frame
import pandas as pd

#Retrieving data from GN
bxd_query = requests.get("http://robot.genenetwork.org/webqtl/main.py?cmd=trait&probeset=11897&db=BXDPublish&format=col") 
print (bxd_query.text)

bxd = pd.read_csv(io.StringIO(bxd_query.text), sep='\t')
print (bxd.head)

#Return data from MPD
cc_resp = requests.get("https://phenome.jax.org/api/pheno/animalvals/43574") 
cc_data = json.loads(cc_resp.text)
cc_animaldata = json_normalize(cc_data['animaldata'])
print (cc_animaldata)

```
